# **Group By**
#### The groupby() function in pandas is used to group data based on one or more columns in a *DataFrame*

```python
df.groupby(['Size'], as_index=False)['Qty'].sum().sort_values(by='Qty',ascending=False)
```

```python
S_Qty=df.groupby(['Size'], as_index=False)['Qty'].sum().sort_values(by='Qty',ascending=False)

sns.barplot(x='Size',y='Qty', data=S_Qty)
```

##### Note: From above Graph you can see that most of the Qty buys M-Size in the sales

```python
data = {'Category': ['Electronics', 'Clothing', 'Electronics', 'Furniture', 'Clothing', 'Furniture', 'Clothing'],
        'Amount': [100, 200, 150, 300, 120, 250, 180]}

df = pd.DataFrame(data)

# Convert 'Category' to string type (as you've done)
df['Category'] = df['Category'].astype(str)

# Count the frequency of each category
category_counts = df['Category'].value_counts()

# Create a pie chart
plt.figure(figsize=(8, 8))  # Adjust the figure size
plt.pie(category_counts, labels=category_counts.index, autopct='%1.1f%%', startangle=140)
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle
plt.title('Distribution of Categories')

# Show the plot
plt.show()
```

```python
df.columns
```

```python
a1 = df['Fulfilment'].value_counts()
fig, ax= plt.subplots()

plt.figure(figsize = (10,6))
ax.pie(a1, labels=a1.index, autopct = '%1.1f%%', radius=0.7, wedgeprops=dict(width=0.6))
ax.set(aspect="equal")

plt.show()
```

###Note: From above chart you can see that most of the Fulfilment are amazon

```python
x_data = df['Category']
y_data = df['Size']

plt.figure(figsize=(8, 8))
plt.scatter(x_data,y_data)
plt.xlabel('Category')
plt.ylabel('Size')
plt.title("Scatter PLot")
plt.show()
```

```python
plt.figure(figsize=(12,6))
sns.countplot(data=df, x='ship-state')
plt.xlabel('ship-date')
plt.ylabel('count')
plt.title('Distribution of state')
plt.xticks(rotation=90)
plt.show()
```

```python
#Top 10 states
top_10_states = df['ship-state'].value_counts().head(10)

plt.figure(figsize = (14,8))
ax= sns.countplot(data=df[df['ship-state'].isin(top_10_states.index)], x='ship-state')
plt.xlabel('ship-state')
plt.ylabel('count')
plt.title("Distribution of State")
for bars in ax.containers:
  ax.bar_label(bars)
plt.show()
```

###Note: From above graph you can see thta most of the buyers are from Maharastra State
